```python
crop_face = face[100:-100, 100:-100]
import numpy as np
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]
y.shape, x.shape
```

```python
centerx, centery = (1200, 600) # центр изображения
mask = ((y - centery)**2 + (x - centerx)**2) > 230**2 # круг
```

```python
face[mask] = 0
plt.imshow(face)  
```

```python
centerx, centery = (1200, 600) # центр изображения
mask = ((y - centery)**2 // 5**2  + (x - centerx)**2 // 8**2 ) > 23**2 # эллипс
```

```python
face[mask] = 0
plt.imshow(face)  
```
