```python
# 23
from collections import Counter
ser = pd.Series(['Apple', 'Orange', 'Plan', 'Python', 'Money'])
cond = ser.map(lambda x: sum([Counter(x.lower()).get(i, 0) for i in ['a','e', 'i', 'o', 'u']]) >= 2)
ser[cond]
```

```python
# 24
import re

emails = pd.Series(['buying books at amazom.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com'])
pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'
r = re.compile(pattern=pattern)
cond = emails.apply(lambda x: bool(r.match(x)))
emails[cond]
```

```python
# 33
def high(x):
    try:
        return float(x)
    except:
        return 0

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv', 
dtype='object')

df.loc[df["medv"].apply(float) < 25, "medv" ] = "Low"
df.loc[df["medv"].apply(high) > 25, "medv" ] = "Hign"
df
```
