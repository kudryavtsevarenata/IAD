```python
# 44

def switch(df, col1, col2):
    colnames = df.columns.tolist()
    ind1, ind2 = colnames.index(col1), colnames.index(col2)
    colnames[ind2], colnames[ind1] = colnames[ind1], colnames[ind2]
    return df[colnames]

df = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))
df = df[list('cbade')]
df = switch(df, 'c', 'a')
df[sorted(df.columns, reverse=True)]
```

```python
# 59
df = pd.DataFrame(np.random.randint(1,100, 40).reshape(10, -1), columns=list('pqrs'), index=list('abcdefghij'))

nearest_row = []
nearest_dist = []

for i, row in df.iterrows():
    curr = row
    rest = df.drop(i)
    dists = {}
    for j, cont in rest.iterrows():
        dists.update({j: round(np.linalg.norm(curr.values - cont.values))})
    nearest_row.append(max(dists, key=dists.get))
    nearest_dist.append(max(dists.values()))
df['nearest_row'] = nearest_row
df['dist'] = nearest_dist
df
```
